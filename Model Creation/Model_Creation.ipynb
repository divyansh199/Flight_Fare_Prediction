### Model Evaluvation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def get_metrics(y_test,y_pred,model_name):
    MSE = mean_squared_error(y_test, y_pred)
    RMSE= np.sqrt(MSE)
    MAE = mean_absolute_error(y_test, y_pred)
    R2 = r2_score(y_test, y_pred)
    score = { 'matices': ['MSE','RMSE','MAE','R2'],
             'values': [MSE,RMSE,MAE,R2]
    }
    return pd.DataFrame(score)
```

```python
pd.options.display.float_format = '{:.12g}'.format
```

```python
get_metrics(y_test, y_pred, "LinearRegression")
```

### HyperParameter Tuning on RandomForestRegressor()

```python
from sklearn.model_selection import RandomizedSearchCV

#create the random grid
random_grid = {
                #number of tree in randomforest
                'n_estimators': [int(x) for x in np.linspace(start = 200 , stop = 2000, num = 10)],

                #number of features to consider at every split
                'max_features': ["auto", "sqrt"],
                
                #maximum number of level allowed in each decision tree
                'max_depth': [int(x) for x in np.linspace(10,120, num=12)],

                #maximun number of sample required to split a node 
                'min_samples_split': [2, 5,10],

                #maximum number of sample required at each leaf node
                'min_samples_leaf': [1,2,4],

                #method of selecting sample for training each tree
                'bootstrap': [True, False] 
              }
```

```python
import time
start_time = time.time()
rfr3 = RandomForestRegressor() 
random_grid = RandomizedSearchCV(estimator = rfr3, param_distributions = random_grid, n_iter =100 , cv = 3, verbose =2, random_state =42, n_jobs = -1)
random_grid.fit(X_train,y_train)
y_pred_rfr3 = random_grid.predict(X_test)
print(f'time taken to traning using randomized search', time.time() - start_time)
```

```python
best_params = random_grid.best_params_
best_params
```

```python
get_metrics(y_test, y_pred_rfr3, "RandomForestRegressor")
```

#### After HyperParameter Tuning R2 score comes out to be 0.84 and Time taken almost 7mins
